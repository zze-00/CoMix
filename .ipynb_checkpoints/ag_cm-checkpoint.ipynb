```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

ab_pairs = [(0.75, 0.75)]

x = np.linspace(0, 1, 1002)[1:-1]

for a, b in ab_pairs:
    print(a, b)
    dist = beta(a, b)
    y = dist.pdf(x)
    plt.plot(x, y, label=r'$\alpha=%.2f,\ \beta=%.2f$' % (a, b))

# 设置标题
plt.title(u'Beta Distribution')
# 设置 x,y 轴取值范围
plt.xlim(0, 1)
plt.ylim(0, 4)
plt.legend()
plt.savefig("./beta.svg", format="svg")
```
